```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from scipy import stats
from sklearn.linear_model import LogisticRegression as lr
from sklearn.metrics import accuracy_score

np.set_printoptions(threshold=np.inf)
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
# load dataset
df = pd.read_csv("train.csv", header=0)
xt=df.drop(["popularity"],axis=1)
yt=df["popularity"]
xte = pd.read_csv("test.csv", header=None)
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(yt)
encoded_Y = encoder.transform(yt)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
print(dummy_y)
```

```python
# define baseline model
# create model
model = Sequential()
model.add(Dense(600, input_dim=6, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(400, activation='relu'))
model.add(Dense(4, activation='softmax'))
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xt, dummy_y, epochs=200, batch_size=10)
scores = model.evaluate(xt, dummy_y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
predictions = model.predict(xte)
fc=predictions.argmax(axis=-1)+1
fc
```

```python
np.savetxt("ans.csv",fc,fmt="%d")
```
